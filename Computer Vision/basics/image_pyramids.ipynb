# Image Pyramiding

## Import all the necessary libraries

```python
import cv2
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read and display an image

```python
#img = cv2.imread("img/self_driving.jpg") # OpenCV image format is`BGR`.
img = mpimg.imread("img/self_driving.jpg") # Matplotlib image format is`RGB`.
fig = plt.figure(figsize=(10, 5))
plt.imshow(img)
```

```python
img.shape
```

## Perform Pyramiding (Down)

```python
scaled_down_1 = cv2.pyrDown(img)
scaled_down_2 = cv2.pyrDown(scaled_down_1)
scaled_down_3 = cv2.pyrDown(scaled_down_2)

print("scaled_down_1: ", scaled_down_1.shape)
print("scaled_down_2: ", scaled_down_2.shape)
print("scaled_down_3: ", scaled_down_3.shape)


fig = plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
plt.imshow(img)
plt.subplot(2,2,2)
plt.imshow(scaled_down_1)
plt.subplot(2,2,3)
plt.imshow(scaled_down_2)
plt.subplot(2,2,4)
plt.imshow(scaled_down_3)
```

## Perform Pyramiding (Up)

```python
scaled_up = cv2.pyrUp(img)

print("scaled_up: ", scaled_up.shape)

fig = plt.figure(figsize=(10, 10))
plt.imshow(scaled_up)
```

# Image Pyramids [Option]



## Theory

Normally, we used to work with an image of constant size.
But on some occasions, we need to work with images of the different resolution of the same image.

- For example:
    while searching for something in an image, like face, we are not sure at what size the object will be present in the image. 

In this case, we need to create a set of images with different resolution and search for the object in all the image.  

These set of images with different resolution are called `Image Pyramids`.

## Image Pyramid

- An image pyramid is a collection of images
- There are tow common kinds of image pyramids:
    - `Gaussain pyramid`: Used to downsample images
    - `Laplacian pyramid`: Used to reconstruct an upsample image from an image lower in th pyramid (with less resolution)
