# Region of Interest (ROI)

## Import all the necessary libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

## Load and display an image

```python
img = cv2.imread("img/lane_1.jpg")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

#img = mpimg.imread("img/lane_1.jpg") # RGB
#plt.imshow(img)
```

## Conver to grayscale

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

## Region of interest selection

```python
height, width = gray.shape

# Select points of the region of interest (ROI)
ROI = np.array([[(100, height),(400, 350), (600, 350), (width, height)]], dtype=np.int32)    

# define a blank image with all zeros (ie: black) 
blank = np.zeros_like(gray)  

print("[origin image] height: {}, width: {}".format(height, width))
print("[blank image]  height: {}, width: {}".format(height, width))     
```

## Perform ROI masking

```python
# Fill the Region of interest with white color (ie: 255)!
mask = cv2.fillPoly(blank, ROI, 255)
    
# Perform bitwise AND operation to select only the region of interest
masked_img = cv2.bitwise_and(gray, mask)

plt.imshow(masked_img, cmap='gray')
```
